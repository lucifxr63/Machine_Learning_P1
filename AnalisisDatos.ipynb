```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo CSV
file_path = 'DataSets/DataSetTokenizados.csv'
data = pd.read_csv(file_path)
data.head()
```

# Análisis de Datos con Histogramas
En esta sección, se analizarán los datos relacionados con Sexo, Edad, CDM, Tipo_GRD, GRD_ y Severidad mediante histogramas. También se calculará la frecuencia de diagnósticos y procedimientos.

```python
# Importar bibliotecas necesarias para visualización
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crear histogramas para Sexo y Edad
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Sexo_bin'], bins=2, kde=False, color='blue')
plt.title('Distribución por Sexo')
plt.xlabel('Sexo (0: Femenino, 1: Masculino)')
plt.ylabel('Frecuencia')
plt.subplot(1, 2, 2)
sns.histplot(data['Edad'], bins=20, kde=True, color='green')
plt.title('Distribución por Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# Crear histogramas para CDM, Tipo_GRD, GRD_ y Severidad (cada uno en una figura separada)
variables = ['CDM', 'Tipo_GRD', 'GRD_', 'Severidad']

for var in variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[var], bins=20, kde=True)
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

# Análisis de Frecuencia de Datos en Columnas de Diagnósticos y Procedimientos
Esta sección analiza la frecuencia con la que encontramos datos válidos en cada columna de diagnósticos y procedimientos.

```python
# Analizar la frecuencia de datos en columnas de diagnósticos
# Contar valores no nulos en cada columna de diagnósticos
diagnosticos_cols = [col for col in data.columns if col.startswith('Diag') and col != 'Diag_Principal' and col != 'Diag_Secundario' and col != 'Diag_Principal_Token' and col != 'Diag_Secundario_Token']
frecuencia_diagnosticos = data[diagnosticos_cols].count().sort_values(ascending=False)



# Calcular el porcentaje de datos respecto al total de registros
total_registros = len(data)
porcentaje_diagnosticos = (frecuencia_diagnosticos / total_registros * 100).round(2)

```

```python
# Analizar la frecuencia de datos en columnas de procedimientos
# Contar valores no nulos en cada columna de procedimientos
procedimientos_cols = [col for col in data.columns if col.startswith('Proced') and col != 'Proced_Principal' and col != 'Proced_Secundario' and col != 'Proced_Principal_Token' and col != 'Proced_Secundario_Token']
frecuencia_procedimientos = data[procedimientos_cols].count().sort_values(ascending=False)

# Calcular el porcentaje de datos respecto al total de registros
porcentaje_procedimientos = (frecuencia_procedimientos / total_registros * 100).round(2)

```

```python
# Visualizar la frecuencia de datos en columnas de diagnósticos
plt.figure(figsize=(14, 8))
ax = frecuencia_diagnosticos.plot(kind='bar', color='skyblue')
plt.title('Frecuencia de Datos en Columnas de Diagnósticos', fontsize=15)
plt.xlabel('Columnas de Diagnósticos', fontsize=12)
plt.ylabel('Número de Registros', fontsize=12)
plt.xticks(rotation=90)

# Añadir etiquetas con los valores
for i, v in enumerate(frecuencia_diagnosticos):
    ax.text(i, v + 100, f'{v}', ha='center', fontsize=9)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualizar la frecuencia de datos en columnas de procedimientos
plt.figure(figsize=(14, 8))
ax = frecuencia_procedimientos.plot(kind='bar', color='lightgreen')
plt.title('Frecuencia de Datos en Columnas de Procedimientos', fontsize=15)
plt.xlabel('Columnas de Procedimientos', fontsize=12)
plt.ylabel('Número de Registros', fontsize=12)
plt.xticks(rotation=90)

# Añadir etiquetas con los valores
for i, v in enumerate(frecuencia_procedimientos):
    ax.text(i, v + 100, f'{v}', ha='center', fontsize=9)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Comparar frecuencias de diagnósticos y procedimientos en un mismo gráfico
# Preparar los datos
max_cols = max(len(diagnosticos_cols), len(procedimientos_cols))
diag_freq = pd.Series(index=range(1, max_cols+1), dtype=float)
proc_freq = pd.Series(index=range(1, max_cols+1), dtype=float)

# Rellenar las series con las frecuencias disponibles
for i, freq in enumerate(frecuencia_diagnosticos, 1):
    diag_freq[i] = freq

for i, freq in enumerate(frecuencia_procedimientos, 1):
    proc_freq[i] = freq

# Crear un DataFrame para facilitar la visualización
freq_comparison = pd.DataFrame({
    'Diagnósticos': diag_freq,
    'Procedimientos': proc_freq
})

# Visualizar la comparación
plt.figure(figsize=(15, 8))
freq_comparison.plot(kind='bar', width=0.7)
plt.title('Comparación de Frecuencias: Diagnósticos vs Procedimientos', fontsize=15)
plt.xlabel('Número de Columna', fontsize=12)
plt.ylabel('Número de Registros', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Calcular estadísticas sobre la cantidad de diagnósticos y procedimientos por paciente
# Contar diagnósticos no nulos por fila
diagnosticos_por_paciente = data[diagnosticos_cols].notna().sum(axis=1)
procedimientos_por_paciente = data[procedimientos_cols].notna().sum(axis=1)

```

```python
# Visualizar la distribución de la cantidad de diagnósticos y procedimientos por paciente
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(diagnosticos_por_paciente, kde=True, color='blue')
plt.title('Distribución de Diagnósticos por Paciente')
plt.xlabel('Número de Diagnósticos')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sns.histplot(procedimientos_por_paciente, kde=True, color='green')
plt.title('Distribución de Procedimientos por Paciente')
plt.xlabel('Número de Procedimientos')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
