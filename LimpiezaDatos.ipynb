```python
import pandas as pd
```

```python
# Intentar cargar el archivo CSV con diferentes configuraciones
try:
    dataset = pd.read_csv('DataSets/dataset_elpino.csv', encoding='utf-8', sep=',')
    print(dataset.columns)
except pd.errors.ParserError:
    print('Error al analizar el archivo CSV. Intentando con otro delimitador o codificación...')
    try:
        dataset = pd.read_csv('dataset_elpino.csv', encoding='latin1', sep=';')
        print(dataset.columns)
    except Exception as e:
        print(f'No se pudo cargar el archivo: {e}')
```

```python
# Renombrar columnas del dataset eliminando espacios, 'Principal (cod+des)', 'Secundario (cod+des)', 'Edad en años' y 'Sexo (Desc)'
columnas_renombradas = {
    col: col.replace(' ', '').replace('Principal(cod+des)', '').replace('Secundario(cod+des)', '').replace('EdadenaÃ±os', 'Edad').replace('Sexo(Desc)', 'Sexo_bin')
    for col in dataset.columns
}
dataset.rename(columns=columnas_renombradas, inplace=True)
print(dataset.columns)
```

```python
# Limpiar datos eliminando espacios y texto después del guion
def limpiar_datos(valor):
    if isinstance(valor, str):
        return valor.split('-')[0].replace(' ', '')
    return valor
dataset = dataset.applymap(limpiar_datos)
print(dataset.head())
```

```python
# Transformar el género a binario: hombre -> 1, mujer -> 0
if 'Sexo_bin' in dataset.columns:
    dataset['Sexo_bin'] = dataset['Sexo_bin'].map({'Hombre': 1, 'Mujer': 0})
    print(dataset['Sexo_bin'].head())
else:
    print('La columna Sexo_bin no existe en el dataset.')
```
